```python
import numpy
import cv2
import os
import random
import matplotlib.pyplot as plt
import pickle
```

```python
DIRECTORY = r'C:\Users\LENOVO\Documents\Python Scripts\datasets\cats-vs-dogs'
```

```python
CATEGORIES = ['cat','dog']
```

```python
IMG_SIZE = 100
data = []
for category in CATEGORIES :
    folder = os.path.join(DIRECTORY,category)
    label = CATEGORIES.index(category)   #this will give label as 0-cat nd 1-dog
    
    for img in os.listdir(folder) :
        img_path = os.path.join(folder,img)
        
        print(img_path)
        
        img_arr = cv2.imread(img_path)
        img_arr = cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        data.append([img_arr,label])
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
print(data[100])
```

```python
X = []
Y = []
for features, labels in data :
    X.append(features)
    Y.append(labels)
```

```python
X = numpy.array(X)
Y = numpy.array(Y)
```

```python
pickle.dump(X, open('X.pkl','wb'))
pickle.dump(Y, open('Y.pkl', 'wb'))
```
